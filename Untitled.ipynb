```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
plt.rcParams['figure.figsize'] = (16, 12)
```

```python
ds = pd.read_csv('edlich-kmeans-A0.csv')
```

```python
ds.max()
```

```python
ds.iloc[:,0]
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(ds.iloc[:,0], ds.iloc[:,1], ds.iloc[:,2])
```

```python
# Initializing KMeans
kmeans = KMeans(n_clusters=5)
# Fitting with inputs
y_pred = kmeans.fit_predict(ds)
```
